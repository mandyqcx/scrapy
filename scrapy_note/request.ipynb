```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://news.sina.com.cn/china/')
res.encoding='utf-8'
soup = BeautifulSoup(res.text,'html.parser')
print(soup)
```

```python
for new in soup.select('.news-1'):
    for li in new.select('li'):
        text=li.text
        a=li.href
        print(text,a)    
```
